# Questão 2 (vale 2 pontos)
Uma empresa espacial está tentando descobrir o ponto mínimo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=7\sin(x)-2\cos(x)$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

Como sempre, você pode usar qualquer ferramenta disponível na internet para lhe auxiliar a resolver este problema. Lembre-se, contudo, que o sympy é uma boa alternativa. 

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio. Lembrem-se estamos limitados as bibliotecas sympy, numpy e matplotlib.

Para resolver esta questão você precisará de, no minimo, 4 casas decimais de precisão depois da vírgula. 

```python
def f(x):
    return 7*np.sin(x) - 2*np.cos(x)

def eagerAdaptado(func, intervalo, passo_inicial, refinamentos, divisoes):
    x_min = intervalo[0]
    x_max = intervalo[1]
    passo = passo_inicial
    pontos_criticos = []

    for _ in range(refinamentos + 1):
        x = x_min
        anterior = func(x)
        crescente = None
        while x < x_max:
            x += passo
            atual = func(x)
            if crescente is None:
                crescente = atual > anterior
            elif crescente and atual < anterior:
                pontos_criticos.append(x - passo / 2)
                crescente = False
            elif not crescente and atual > anterior:
                pontos_criticos.append(x - passo / 2)
                crescente = True
            anterior = atual
        passo /= divisoes

    return pontos_criticos

intervalo = (-10, 10)
passo_inicial = 1
refinamentos = 3
divisoes = 100
pontos_criticos = eagerAdaptado(f, intervalo, passo_inicial, refinamentos, divisoes)

print("Pontos críticos encontrados:")
for x in pontos_criticos:
    print("x = {:.4f}".format(x))



display(Math(r"7sin(x)-2cos(x) = 0"))
display(Math(r"7sin(x) = 2cos(x)"))
display(Math(r"7tan(x) = 2"))
display(Math(r"tan(x) = \frac{2}{7}"))
display(Math(r"x = arctan(\frac{2}{7}) + \pi n "))
display(Math(r"x = 0.27829... + \pi n "))
print("\n")
display(Math(r"f(x) = 7sin(0) = 2cos(0)"))
display(Math(r"f(x) = -2"))
display(Math(r"f(x) = 7sin(0.1) = 2cos(0.1)"))
display(Math(r"f(x) = -1.29117..."))
```

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+2y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return np.sin(x*y) + np.cos(x + 2*y)

# Create a 2D grid of points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)


Z = f(X, Y)


max_idx = np.unravel_index(np.argmax(Z), Z.shape)
min_idx = np.unravel_index(np.argmin(Z), Z.shape)


plt.imshow(Z, cmap='viridis', extent=[-5, 5, -5, 5])
plt.colorbar()
plt.title('f(x,y) = sin(xy) + cos(x+2y)')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x[max_idx[1]], y[max_idx[0]], marker='o', color='red', s=100)
plt.scatter(x[min_idx[1]], y[min_idx[0]], marker='o', color='blue', s=100)
plt.show()
```
