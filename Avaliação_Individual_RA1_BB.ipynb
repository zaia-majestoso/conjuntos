# Questão 1: 

Um país tem uma população inicial de 10 milhões de habitantes e cresce anualmente a uma taxa de 2,5%. Faça um gráfico que mostre como a população desse país varia com o tempo e determine quantos habitantes ele terá após 20 anos. Utilize a equação $P = P0 * (1 + r)^t$, onde $P0$ é a população inicial, $r$ é a taxa de crescimento anual (em decimal) e $t$ é o tempo decorrido em anos. 

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import math

x = sp.symbols('x')
xp = np.linspace(0, 20, 200)
yp = 10000000 * ((1 + 0.025)**xp)
result = 10000000 * ((1 + 0.025)**20)
plt.plot(xp, yp)
plt.xlabel('anos')
plt.ylabel('população')
plt.title(result)

```
